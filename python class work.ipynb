```python
#adding a legend to a graph 

from matplotlib import pyplot as plt

fig=plt.figure()
axes1=fig.add_axes([0.1,0.2,0.8,0.9])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'green')
axes2.plot(y,x,'red')
#adding a legend to a graph

axes1.plot(x,x**2, label='x squared')
axes2.plot(y,y**3, label='y cubed')
axes1.legend()
```
